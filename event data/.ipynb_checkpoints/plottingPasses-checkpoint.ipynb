```python
import numpy as np
import matplotlib.pyplot as plt
from mplsoccer import Pitch, Sbopen
```

```python
parser = Sbopen()
df, related, freeze, tactics = parser.event(69301)
df.info()
```

```python
team1, team2 = df.team_name.unique()
team1, team2
```

```python
event_types = df.type_name.unique()
event_types
```

Create a dataframe of passes which contains all the passes in the match

```python
passes = df[df.type_name == 'Pass']
passes
```

Plot the start point of every Sweden pass. Attacking left to right.

```python
pitch = Pitch(line_color='black')
fig, ax = pitch.grid(grid_height=0.9, axis=False, title_height=0.06, endnote_height=0.04, title_space=0, endnote_space=0)
for i, row in passes.iterrows():
    if row.team_name == team2 and row.sub_type_name != 'Throw-in':
        x = row['x']
        y = row['y']
        pitch.scatter(x,y,alpha=0.7,s=100,color='blue',ax=ax['pitch'])
fig.suptitle('start point of every sweden pass\
 attacking left to right',fontsize=30)
plt.show()
```

Plot only passes made by Caroline Seger (she is Sara Caroline Seger in the database)

```python
mask_seger = (df.type_name == 'Pass') & (df.sub_type_name != 'Throw_in') & (df.player_name == 'Sara Caroline Seger')
df_seger = df.loc[mask_seger, ['x','y','end_x','end_y']]
pitch2 = Pitch(line_color='black')
fig, ax = pitch.grid(grid_height=0.9,axis=False,title_height=0.06,endnote_height=0.04,title_space=0,endnote_space=0)
for i, row in df_seger.iterrows():
    x=row['x']
    y=row['y']
    pitch2.scatter(x,y,alpha=0.5,s=100,color='red',ax=ax['pitch'])
fig.suptitle('passes made by caroline seger when attacking from left to right',fontsize=30)
plt.show()
```

Plot arrows to show where the passes went to.

```python
pitch3 = Pitch(line_color='black')
fig, ax = pitch.grid(grid_height=0.9,axis=False,title_height=0.06,endnote_height=0.04,title_space=0,endnote_space=0)
pitch3.arrows(df_seger.x,df_seger.y,df_seger.end_x,df_seger.end_y,color='blue',ax=ax['pitch'])
pitch3.scatter(df_seger.x,df_seger.y,alpha=0.5,s=100,color='red',ax=ax['pitch'])
fig.suptitle('all caroline segers pass with a starting position in red and an arrow in blue to show the direction and distance of the pass', fontsize=15)
plt.show()
```
